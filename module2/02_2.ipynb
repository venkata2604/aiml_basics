```python
# Categorical values encoded as dummies

df = pd.read_csv("https://data.heatonresearch.com/data/t81-558/jh-simple-dataset.csv",
    na_values=["NA", "?"],)

display(df)
```

```python
df["area"].unique()
```

```python
one_h_encoding = pd.get_dummies(df["area"], prefix="area", dtype=int)
one_h_encoding
```

```python
df = pd.concat([df, one_h_encoding], axis=1)
```

```python
df
```

```python
df1 = df.drop("area", axis=1)
```
